```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read CSV
data = pd.read_csv("output/heisenberg_regular.csv")

# Extract columns
T = data['T']
E = data['E_avg']
M = data['M_avg_mag']
chi = data['chi']
cv = data['cv']

# Plot settings
plt.figure(figsize=(12, 10))

# Energy
plt.subplot(2, 2, 1)
plt.plot(T, E, 'b.-', label='Energy')
plt.vlines(1.443, -3, 0, linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Average Energy per Spin')
plt.title('Energy vs Temperature')
plt.grid(True)
plt.legend()

# Magnetization
plt.subplot(2, 2, 2)
plt.plot(T, M, 'r.-', label='Magentization')
plt.vlines(1.443, 0, 1.1 * max(M), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Average Magnetization per Spin')
plt.title('Magnetization vs Temperature')
plt.grid(True)
plt.legend()

# Susceptibility
plt.subplot(2, 2, 3)
plt.plot(T, chi, 'g.-', label='Susceptibility')
plt.vlines(1.443, 0, 1.1 * max(chi), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Magnetic Susceptibility ()')
plt.title('Susceptibility vs Temperature')
plt.grid(True)
plt.legend()

# Specific Heat
plt.subplot(2, 2, 4)
plt.plot(T, cv, 'm.-', label='Specific Heat')
plt.vlines(1.443, 0, 1.1 * max(cv), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Specific Heat (Cv)')
plt.title('Specific Heat vs Temperature')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig(r"plots/heisenberg_parallel.pdf")
plt.show()

print(T[np.argmax(chi)], T[np.argmax(cv)])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read CSV files
data_ml_batch = pd.read_csv("output/heisenberg_ml_batch.csv")
data_ml = pd.read_csv("output/heisenberg_ml.csv")
data_parallel = pd.read_csv("output/heisenberg_regular.csv")

# Extract columns from all datasets
T_ml_batch = data_ml_batch['T']
E_ml_batch = data_ml_batch['E_avg']
M_ml_batch = data_ml_batch['M_avg_mag']
chi_ml_batch = data_ml_batch['chi']
cv_ml_batch = data_ml_batch['cv']

T_ml = data_ml['T']
E_ml = data_ml['E_avg']
M_ml = data_ml['M_avg_mag']
chi_ml = data_ml['chi']
cv_ml = data_ml['cv']

T_parallel = data_parallel['T']
E_parallel = data_parallel['E_avg']
M_parallel = data_parallel['M_avg_mag']
chi_parallel = data_parallel['chi']
cv_parallel = data_parallel['cv']

# Plot settings
plt.figure(figsize=(12, 10))

# Plot Energy
plt.subplot(2, 2, 1)
plt.plot(T_ml_batch, E_ml_batch, 'b.-', label='ML-Batch')  # Renamed label
plt.plot(T_ml, E_ml, 'r.-', label='ML')  # Renamed label
plt.plot(T_parallel, E_parallel, 'g.-', label='Regular')  # Renamed label
plt.vlines(1.443, min(min(E_ml_batch), min(E_ml), min(E_parallel)), max(max(E_ml_batch), max(E_ml), max(E_parallel)), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Average Energy per Spin')
plt.title('Energy vs Temperature')
plt.grid(True)
plt.legend()

# Plot Magnetization
plt.subplot(2, 2, 2)
plt.plot(T_ml_batch, M_ml_batch, 'b.-', label='ML-Batch')  # Renamed label
plt.plot(T_ml, M_ml, 'r.-', label='ML')  # Renamed label
plt.plot(T_parallel, M_parallel, 'g.-', label='Regular')  # Renamed label
plt.vlines(1.443, 0, 1.1 * max(max(M_ml_batch), max(M_ml), max(M_parallel)), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Average Magnetization per Spin')
plt.title('Magnetization vs Temperature')
plt.grid(True)
plt.legend()

# Plot Susceptibility
plt.subplot(2, 2, 3)
plt.plot(T_ml_batch, chi_ml_batch, 'b.-', label='ML-Batch')  # Renamed label
plt.plot(T_ml, chi_ml, 'r.-', label='ML')  # Renamed label
plt.plot(T_parallel, chi_parallel, 'g.-', label='Regular')  # Renamed label
plt.vlines(1.443, 0, 1.1 * max(max(chi_ml_batch), max(chi_ml), max(chi_parallel)), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Magnetic Susceptibility ()')
plt.title('Susceptibility vs Temperature')
plt.grid(True)
plt.legend()

# Plot Specific Heat
plt.subplot(2, 2, 4)
plt.plot(T_ml_batch, cv_ml_batch, 'b.-', label='ML-Batch')  # Renamed label
plt.plot(T_ml, cv_ml, 'r.-', label='ML')  # Renamed label
plt.plot(T_parallel, cv_parallel, 'g.-', label='Regular')  # Renamed label
plt.vlines(1.443, 0, 1.1 * max(max(cv_ml_batch), max(cv_ml), max(cv_parallel)), linestyle='dashed', color='black', label=r'Theoretical $T_c$')
plt.xlabel('Temperature (T)')
plt.ylabel('Specific Heat (Cv)')
plt.title('Specific Heat vs Temperature')
plt.grid(True)
plt.legend()

# Adjust layout and save
plt.tight_layout()
plt.savefig(r"plots/combined_properties_mrt.png")
plt.show()

print(r"Tc for Regular step function: Tc =", T_parallel[np.argmax(chi_parallel)], T_parallel[np.argmax(cv_parallel)])
print(r"Tc for ML step function: Tc =", T_parallel[np.argmax(chi_ml)], T_parallel[np.argmax(cv_ml)])
print(r"Tc for ML-Batch step function: Tc =", T_parallel[np.argmax(chi_ml_batch)], T_parallel[np.argmax(cv_ml_batch)])
```
